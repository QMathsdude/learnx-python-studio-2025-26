**`while` loops**

````python
# A while loop repeatedly executes its body as long as the condition is true
n = 0
while n < 5:
    print(n, end=" ") # Instead of printing onto a new line, print with spaces instead
    n += 1
print()

# WARNING: A while True loop will loop infinitely, so include a break statement
# as below to exit the loop at the break statement
# break statements exit the innermost loop the code is currently at
n = 1
while True:
    print(n, end=" ")
    if n % 7 == 0:
        break
    n += 4
print()

# the continue statement skips the loop to the next iteration,
# ignoring any more lines of code below it
# Print multiples of 2 that aren't divisible by 3
n = 0
while n < 20:
    n += 2
    if n % 3 == 0:
        continue # Continue to next iteration to prevent the print
    print(n, end=" ")
````

**`for` loops**

````python
# With for loops, you can iterate over each item of a list

lst = [837, 128, 124, 12, 93, 192, 543, 798]

# Print out each item of a list
for item in lst:
    print("Item:", item)

# Similarly, we can use break or continue in a for loop
# Print out all odd numbers in lst
for item in lst:
    if item % 2 == 0:
        continue
    print("Item (Odd):", item)
````

````python
# We don't necessarily need to iterate over lists
# For example, there is a very useful object called `range`s
# Which generates a sequence of numbers

for i in range(20): # Range does not include 20!
    print(i, end=' ')
print()

# Print integers from 1 to 200 which are multiples of 7
# and one more than a multiple of 4
for i in range(1, 201): # Range specified as [1, 201)
    if i % 7 == 0 and i % 4 == 1:
        print(i, end=' ')
print()
````

### Functions

Functions are basically named blocks of code that can be referred back to. They can take in multiple parameters
and return an output value with the `return` statement. An important thing to note about `return`
is that it terminates the function's execution immediately and proceeding lines are ignored.

````python
def square(x):
    return x * x

print("13 squared is:", square(13))
print("1728 squared is:", square(1728))
````

Functions can be used to define complex computations and rewrite code more cleanly.
For example, we can define a function that returns `True` when the input integer is prime,
`False` otherwise. Then we can use the function to print out all prime numbers up to a certain range:

````python
# VERY inefficient implementation for the sake of simplicity
def is_prime(n):
    if n < 2: # 0 and 1 (and negatives) are not primes
        return False
    for d in range(2, n): # Iterate d from 2 up to, not including the input, n
        if n % d == 0: # If the input is divisible by d
            return False # Then the number is definitely not prime
    # Once the loop reaches here,
    # It means the input n is not divisible by any number less than it
    # Thus it is prime
    return True

# Print all prime numbers under 100
for i in range(100):
    if is_prime(i):
        print(i, end=' ')
````

Functions can also refer to themselves, this is called recursion.
For example, the factorial has the following properties:
$$
n!\ =\ n(n - 1)!
$$$$
0!\ =\ 1
$$
This could be translated directly to Python code:

````python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print("16! =", factorial(16))
````

Similarly to `while True` loops, if we don't define a base case that causes the chain to terminate,
then the function will just call itself infinitely and crash the program.
So exercise caution when recursing (it's a very powerful way to simplify certain computations!).

If you want to return more than one value in a function, return them in a tuple. Then, you can get both the values at once using a special syntax.

````python
from math import cos, sin, pi

def polar_to_cartesian(r, theta):
    return r * cos(theta), r * sin(theta)

# x and y will get automatically assigned because polar_to_cartesian
# returns a 2-tuple
x, y = polar_to_cartesian(2, pi / 2)
print(x, y)
````

### Iterators

If you try print a `range`, you would get:

````python
print(range(10))
````

You might've expected a `list` or something similar. But creating a list just to count values is very
inefficient. Instead, Python's `range` has an internal counter. Each time you want a new value from
the range, it just increments the counter and returns its value.

Anything that represents a sequence like this is an iterator. Above, when I dropped the square brackets,
it creates a generator. Instead of creating a list with all the values in it like a list
comprehension should, it actually just performs the computation on an element of the list when as we iterate
through.

Delaying computation like this to save space is common among iterators, and is generally called "lazy evaluation".
The advantage is obviously with saving memory, but also to prevent unnecessary computation. For example with above,
the `and` function will stop when it encounters a `False` value because the result is guaranteed to be `False` in
that case. Using generators means that any computation after the `False` one won't be performed.

Aside from generators, there are many functions that create unique iterators with many uses.

````python
def ordinal_suffix(n):
    rightmost_digit = n % 10
    if rightmost_digit == 1:
        return 'st'
    elif rightmost_digit == 2:
        return 'nd'
    elif rightmost_digit == 3:
        return 'rd'
    else:
        return 'th'

# If you want the index of an item when iterating through a list.
wordlist = ['blue', 'cricket', 'sundew', 'snowfall', 'luna']
for i, word in enumerate(wordlist):
    num = i + 1
    print(f'{num}{ordinal_suffix(num)} word: {word}')
````

````python
# The zip function combines two iterators by associating
# 1st item of one to 1st of other, 2nd item of one to 2nd of other...
# like a zipper closing two sides of the zip together.
# Each element of the output is a tuple
xlst = [1, 2, 3]
ylst = [4, 5, 6]
for x, y in zip(xlst, ylst):
    print(x, y)
print()

xvals = [4.5, 7.6, 10.1, 15.1]
for x, y in zip(xvals, (x*x for x in xvals)):
    print(f'{x}^2 = {y}')
````

````python
from itertools import product, permutations, combinations

# Cartesian product
xlst = [1, 2, 3]
ylst = [4, 5, 6]
for y, x in product(ylst, xlst):
    print(f'({x}, {y})')
print()

# All possible permutations
for perm in permutations(xlst):
    print(perm)
print()

# All possible combinations of 3 elements
for comb in combinations([1, 2, 3, 4, 5], 3):
    print(comb)
````

### Lambdas and First Class Functions

If you've ever forgotten to write brackets `()` on functions, you might've gotten something like this:

````python
def my_very_cool_func():
    ...

print(my_very_cool_func)
````

If you exepcted it to error, it might surprise you that there is any output at all.

This happens because functions themselves are also objects like `str`s and `int`s,
and can be stored in variables and such. It's just that functions have a syntax that
is used to call them.

You know how to declare functions using `def`, but simple functions that only return
a value can be specified using lambda syntax.

````python
square = lambda x: x*x

print(square(13))
print(square(1728))
````

````python
def repeat_func(func, n):
    # We define the result function...
    def repeated(x):
        result = x
        for i in range(n):
            result = func(result)
        return result
    # ... and return it here
    return repeated

from math import sin

sine_thrice = repeat_func(sin, 3)
print('sin(sin(sin(1))) =', sine_thrice(1))
````

A use for this is with `functools.reduce`. In short, it applies the function to each consecutive
element, accumulating the result from left to right. e.g.:
$$\text{reduce}(f, [a, b, c, d, e])\ =\ f(f(f(f(a, b), c), d), e)$$
To see a more obvious use, consider reducing with the multiplication operator (or any binary operator), then:
$$\text{reduce}(*, [a, b, c, d, e])\ =\ a * b * c * d * e$$

````python
from functools import reduce

def product(lst):
    return reduce(lambda a, b: a * b, lst)

def factorial(n):
    return product(range(1, n + 1))

print("Product:", product([8, 2, 3, 17]))
print("16! =", factorial(16))
````

### Miscellaneous Features

**Dynamic Arguments**

Function arguments can include `*args` and/or `**kwargs` that allow it to take a variable amount of arguments.
For example:

````python
# Function normally takes only 2 arguments (ignoring args, kwargs)
def func(arg1, arg2, *args, **kwargs):
    print(f'{arg1 = }; {arg2 = }; {args = }; {kwargs = }')

# Calling function normally
func(1, 2)
# Excess parameters get passed over to *args as a tuple
func(1, 2, 3, 4)
# Specifying so-called keyword-args (kwargs) as such
# causes it to pass over to *kwargs as a dict
func(1, 2, x=1, the_value='Hello, World!')
````

Using this feature is somewhat rare in scripts but common in library functions that your scripts would use.
So take note of this on later lessons.

Furthermore, there is a converse construct for passing arguments to functions.

````python
from math import cos, sin, pi

def sph_to_rect(r, theta, phi):
    return r * cos(phi) * sin(theta), r * sin(phi) * sin(theta), r * cos(theta)

# Define a point in spherical coordinates with a 3-tuple
pt = (2.5, pi / 2, pi / 3)

# We want to convert to cartesian, but the function accepts 3 arguments separately!
# You can do this:
print(sph_to_rect(pt[0], pt[1], pt[2]))

# But this is equivalent and much more convenient (also works with lists)
print(sph_to_rect(*pt))
````

### Reading Error Messages

Sometimes, you make a mistake in the code and Python errors as a result.

````python
import numpy as np

sum_of_ints = np.sum(n for n in range(101))
print(sum_of_ints)
````

```
TypeError: Calling np.sum(generator) is deprecated.Use np.sum(np.fromiter(generator)) or the python sum builtin instead.
```
As you can see, there are already recommendations, but let's just make a web search to know our problem.
Since we are using the `numpy` library, let's search something like: [numpy generator deprecated](https://www.google.com/search?q=numpy+generator+deprecated&udm=14).

For me, the first result is [an old reddit thread](https://www.reddit.com/r/Numpy/comments/ottyrd/deprecationwarning_calling_npsumgenerator_is/) that shows the same problem.

that provide very helpful insight into the problem by real actual humans.

Now let's follow the above recommendations, using the Python `sum` builtin:

````python
sum_of_ints = sum(n for n in range(101))
print(sum_of_ints)
````

Nice! It works. What about the other recommendation though?

````python
import numpy as np

sum_of_ints = np.sum(
    np.fromiter(n for n in range(101))
)
print(sum_of_ints)
````

Hmm... We get an error again.
```
TypeError: fromiter() missing required argument 'dtype' (pos 2)
```
This means that we didn't pass the 2nd argument into the function, so let's do that:

````python
import numpy as np

sum_of_ints = np.sum(
    np.fromiter(n for n in range(101), dtype=np.int64)
)
print(sum_of_ints)
````

Oh right, generators need brackets around them (unless they're the lone argument to a function).

````python
import numpy as np

sum_of_ints = np.sum(
    np.fromiter((n for n in range(101)), dtype=np.int64)
)
print(sum_of_ints)
````

This works now too!

````python
import matplotlib.pyplot as plt
from math import sqrt,pi
GM=11506.1*1.
m=0.85
d0=50.
v0=12.
def agrav(x,y):
    xn=-x/sqrt(x*x+y*y)
    yn=-y/sqrt(x*x+y*y)
    g=GM*m/(x*x+y*y)
    return g*xn,g*yn
dt=0.05
x,y=d0,0.
vx,vy=0.,v0
ax0,ay0=grav(x, y)
xs=[x]
ys=[y]
while True:
    x+=vx*dt+0.5*ax0*dt*dt
    y+=vy*dt+0.5*ay0*dt*dt
    xs.append(x)
    ys.append(y)
    ax1,ay1=agrav(x,y)
    vx+=0.5*(ax0+ax1)*dt
    vy+=0.5*(ay0+ay1)*dt
    ax0,ay0=ax1,ay1
    if ys[-1]<0 and ys[-2]>=0:
        break
n=len(xs)-1+ys[-1]/(ys[-2]-ys[-1])
t=n*dt*2
o=(ys[-1]*xs[-2]-ys[-2]*xs[-1])/(ys[-2]-ys[-1])
a=(o+d0)/2
est=0.5*t*sqrt(GM/a**3)
print('estimate:',est)
print('err:',abs(pi-est))
plt.gca().set_aspect('equal')
plt.plot(xs, ys, 'r--')
plt.plot([0],[0],'bx')
plt.grid()
plt.show()
````
